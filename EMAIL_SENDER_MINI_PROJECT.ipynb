THIS IS A PROGRAMME TO SEND BIRTHDAY WISHES TO YOUR FRIENDS ON THEIR BIRTHDAY.......


```python
import pandas as pd
import datetime as dt
import smtplib
GMAIL_ID='xxxx@gmail.com';# sender email id .....
GMAIL_PSWD=''; # sender id paswd....
def sendemail(to,subject,msg):# function to send mail....
   print(f"send email to {to} regarding {subject}")
   s=smtplib.SMTP('smtp.gmail.com',587)
   s.starttls();
   s.login(GMAIL_ID,GMAIL_PSWD)
   s.sendmail(GMAIL_ID,to,f"subject : {subject}\n\n {msg}");
   s.quit(); # email sent ..
 # store name ,birthday Date ,email address and dialogue of friends .....
 # in data enter email id's of person's to whom you want to send Wishes . ......
data={  'name':["RIDHAM","Tushar"],'Birthday':["27-01","27-04"],'dialogue':["Happy Birthday Ridham","Hello Tushar"],'email':['XXXXXXXX@gmail.com','XXXXX@gmail.com'],'year':[2023,2023]
     }
df=pd.DataFrame(data)

today=dt.datetime.now().strftime("%d-%m");# today's date
yearnow=dt.datetime.now().strftime("%y");
for i,ind in df.iterrows():
  bd=ind['Birthday'];
  if (bd==today and ind['year']!=str(yearnow)):# if ind[year]==yearnow then todays mail allready sent....
    sendemail(ind['email'],"happy birthday",ind['dialogue']);
    df.loc[i,'year']= str(20)+yearnow;# replace old year with current year

df
```
